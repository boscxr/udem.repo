### Introducción


En esta laboratorio, nosotros trabajaremos con imagenes a color como en blanco y negro. La diferencia entre estas dos es el número de canales que cada una tiene. Una imagen en blanco y negro puede ser representada como una matriz bidimensional MxN, donde M y N representa la resolución en largo y ancho. Para una imagen a color, eso puede ser instanciado con una matriz tridimensional MxNxC, donde el parametro extra "C", representa el número de canales. Se trabajará con pixeles, ya sea en el encontrar su valor como también el modificarlo. Con un arreglo de pixeles podemos generar regiones de interés y esto considerando los tiempos de procesamiento que realiza el CPU.

### c. Retrieving intensity values of pixels
Para recuperar los valores de intensidad de un pixel en específico, tu necesitas especificar la fila y la columna del pixel de interés. Por ejemplo en el código que se encuentra debajo, estamos interesados en el pixel encontrado en las coordenadas (100,100), esto se realiza declarando ```img_colour[100,100]```, el valor del pixel de cada canal es arrojado, en este caso es de [157,151,151]. Existe otra manera de leer valores de pixeles, image pre-built method item(). Se utiliza con ```img_colout.item(100,100,0)```; en este caso tu leerías el canal rojo en la coordenada (100,100), para leer el valor del canal verde y rojo se cambia el argumento a ```img_colour.item(100, 100, 1)``` o ```img_colour.item(100,100,2)```. Para modificar los valores de los pixeles, usas la función ```img_colour[100,100]=[0,0,0]``` para setear los canales rojo, verde y azul (0,0,0), lo que hará que ese pixel sea negro. Aqui tambien podemos usar itemset() como ```img_colour.item(100,100,0)```. Esto hará el canal rojo a 0. Para setear el canal verde y azul a 0. También se puede usando la expresión ```img_colour.item(100,100,1)``` y ```img_colour.item(100,100,2)```. El código que se encuentra a continuación despliega la imagen a color y en blanco y negro, y después las vuelve a imprimir pero con el pixel (100,100) modificado en ambas imagenes.



````python
# read image
image_name = 'figs/vehicular_traffic.jpg'
img_colour = cv2.imread(image_name, cv2.IMREAD_COLOR)
img_colour = cv2.cvtColor(img_colour, cv2.COLOR_BGR2RGB)

# verify that image exists
if img_colour is None:
    print('ERROR: image ', image_name, 'could not be read')
    exit()

# convert the input colour image into a grayscale image
img_greyscale = cv2.cvtColor(img_colour, cv2.COLOR_BGR2GRAY)

# visualise image using matplotlib
plt.figure(1)
plt.imshow(img_colour)
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# visualise image using matplotlib
plt.figure(2)
plt.imshow(img_greyscale, cmap='gray')
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# retrieve pixel intensity value at given row/col of colour image
row, col = 100, 100

# ----------------------- READ PIXEL VALUE ----------------------------- #
print('COLOUR IMAGE:')
print('\n\tpixel intensity value at', '(', row, ',', col, '): ', img_colour[row, col])
print('\n\tAccessing pixel values using image array')
print('\t\tRED pixel intensity value at', '(', row, ',', col, '): ', img_colour[row, col, 0])
print('\t\tGREEN pixel intensity value at', '(', row, ',', col, '): ', img_colour[row, col, 1])
print('\t\tBLUE pixel intensity value at', '(', row, ',', col, '): ', img_colour[row, col, 2])
print('\n\tAccessing pixel values using pre-built method item()')
print('\t\tRED pixel intensity value at', '(', row, ',', col, '): ', img_colour.item(row, col, 0))
print('\t\tGREEN pixel intensity value at', '(', row, ',', col, '): ', img_colour.item(row, col, 1))
print('\t\tBLUE pixel intensity value at', '(', row, ',', col, '): ', img_colour.item(row, col, 2))

print('\nGREYSCALE IMAGE:')
print('\n\tpixel intensity value at', '(', row, ',', col, '): ', img_greyscale[row, col])
print('\n\tAccessing pixel values using image array')
print('\t\tgrey pixel intensity value at', '(', row, ',', col, '): ', img_greyscale[row, col])
print('\n\tAccessing pixel values using pre-built method item()')
print('\t\tgrey pixel intensity value at', '(', row, ',', col, '): ', img_greyscale.item(row, col))

# -------------------- MODIFY PIXEL VALUE ------------------------------ #
# modify pixel intensity values in colour image
img_colour[row, col] = [0, 0, 0]

# uncomment the lines below if you want to modify the pixel value using the pre-built method set()
#img_colour.itemset((row,col,0),0)
#img_colour.itemset((row,col,1),0)
#img_colour.itemset((row,col,2),0)

# visualise image using matplotlib
plt.figure(3)
plt.imshow(img_colour)
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')


# modify pixel intensity values in greyscale image
img_greyscale[row, col] = 0

# uncomment the lines below if you want to modify the pixel value using the pre-built method itemset()
#img_greyscale.itemset((row,col),0)

# visualise image using matplotlib
plt.figure(4)
plt.imshow(img_greyscale,cmap='gray')
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# visualise figures
plt.show()
````

### d. Creating a Region Of Interest

Dependiendo de la aplicación de la visión computacional, tu muchas veces quieres extraer una region de interés (ROI en ingles), esto para acelerar la compilación del código, o para limitar el area en la que se corre un proceso. Supongamos que para una aplicación analítica, se necesita capturar la cantidad de carros que pasan por un determinado carril, si tu logras extrar una región de interés con una parte de la imagen que corresponda a ese carril, tu puedes correr tu código en esa sub-imagen.

El código que se encuentra a continuación nos enseña como tu puedes extrar ROI, que corresponde a el carro rojo. Después el carro rojo (RIO) es pegado en la imagen a color original para crear una imagen completamente nueva. 

- ```img_roi=img_colour[550:700, 630:750]``` así declaramos nuestra región de interés
- ```img_colour[570:720, 880:1000]=img_roi``` así pegamos nuestro ROI en una nueva imagen



````python
# import required libraries
import matplotlib.pyplot as plt
import cv2

# read image
image_name = 'figs/vehicular_traffic.jpg'
img_colour = cv2.imread(image_name, cv2.IMREAD_COLOR)
img_colour = cv2.cvtColor(img_colour, cv2.COLOR_BGR2RGB)

# verify that image exists
if img_colour is None:
    print('ERROR: image ', image_name, 'could not be read')
    exit()


# visualise image using matplotlib
plt.figure(1)
plt.imshow(img_colour)
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')


# create a region of interest
# NOTE THE FOLLOWING:
# FIRST RANGE REPRESENTS COORDINATES FOR THE Y-AXIS
# SECOND RANGE REPRESENTS COORDINATES FOR THE X-AXIS
img_roi = img_colour[550:700, 630:750]

# visualise region of interest ROI in a new figure
plt.figure(2)
plt.imshow(img_roi)
plt.title('OUTPUT IMAGE: Region Of Interest')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# overlay roi on colour image
img_colour[570:720, 880:1000] = img_roi
#img_colour[550:700, 450:570] = img_roi

# visualise region of interest ROI on colour image
plt.figure(3)
plt.imshow(img_colour)
plt.title('roi image overlaid on colour image')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# display figures
plt.show()
````

### e. Splitting and merging image channels

Sometimes, you will need to work with each image channel separately. This is the case when detecting and tracking lanes on a highway. Hence, in this section of the lab you will learn how to do so and conversely, how to merge individual channels to build an RGB image. The two methods to accomplish this are cv2.split() and cv2.merge(). The code below shows you how to use these two method to split or merge channels. Copy this code into a Python script and run it.

````python
# import required libraries
import numpy as numpy
import matplotlib.pyplot as plt
import cv2

# read image
image_name = 'figs/vehicular_traffic.jpg'
img_colour = cv2.imread(image_name, cv2.IMREAD_COLOR)
img_colour = cv2.cvtColor(img_colour, cv2.COLOR_BGR2RGB)

# verify that image exists
if img_colour is None:
    print('ERROR: image ', image_name, 'could not be read')
    exit()

# visualise image using matplotlib
plt.figure(1)
plt.imshow(img_colour)
plt.title('INPUT IMAGE: COLOUR')
plt.xlabel('x-resolution')
plt.ylabel('y-resolution')

# split colour image channels
r_channel,g_channel,b_channel = cv2.split(img_colour)

# visualise each channel
plt.figure(2)
plt.imshow(r_channel,cmap='gray')
plt.title('RED channel')
plt.figure(3)
plt.imshow(g_channel,cmap='gray')
plt.title('GREEN channel')
plt.figure(4)
plt.imshow(b_channel,cmap='gray')
plt.title('BLUE channel')

# merge channels
img_original = cv2.merge((r_channel, g_channel, b_channel))

plt.figure(5)
plt.imshow(img_original)
plt.title('Merged channels')

# set blue channel to zero
b_channel[:,:] = 0

# merge channels after setting blue channel to zero
img_bzero = cv2.merge((r_channel, g_channel, b_channel))

plt.figure(6)
plt.imshow(img_bzero)
plt.title('Merged channels after setting BLUE channel to zero')

# display figures
plt.show()
````

En las seis imagenes anteriores se pueden observar, la diferencia de tonalidades entre los tres canales de RGB (Red,Green,Blue). Al momento de separar un canal de los otros dos, lógicamente se hace blanco y negro, porque las imagenes con un solo canal son en blanco y negro. En el canal rojo, se diferencía que el carro rojo se vuelve blanco a diferencia de los otros canales, esto debido a que los valores del pixel en el canal rojo son valores muy altos, esto al hacerlo a un solo canal se vuelve blanco.

En la última imagen se observa una imagen amarilla, esto es fácil de analizar puesto que se le quito el canal azul, dejando solamente el rojo y el verde funcionando. Al momento de fusionar el color rojo con el verde, se genera un amarillo que es lo que se visualiza en la imagen del final.
